```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Load data (no gravity)

```python
path = 'data2/run1/'
data0 = pd.read_csv(path+"run1_0.csv")
data1 = pd.read_csv(path+"run1_5.csv")
data2 = pd.read_csv(path+"run1_10.csv")
data3 = pd.read_csv(path+"run1_15.csv")
data4 = pd.read_csv(path+"run1_20.csv")
```

### Plotting density and velocity



```python
data0
```

```python
fig, ax = plt.subplots(5,1, sharex=True, figsize=(8, 8))

x = np.arange(0, 64, 1)
print(len(x))

rho_0 = data0['rho']
rho_5 = data1['rho']
rho_10 = data2['rho']
rho_15 = data3['rho']
rho_20 = data4['rho']

ax[0].plot(x, rho_0, label='t = 0')
ax[1].plot(x, rho_5, label='t = 5')
ax[2].plot(x, rho_10, label='t = 10')
ax[3].plot(x, rho_15, label='t = 15')
ax[4].plot(x, rho_20, label='t = 20')

ax[4].set_xlabel("Cell number", fontsize=15)

ax[0].set_ylabel('Density', fontsize=15)
ax[1].set_ylabel('Density', fontsize=15)
ax[2].set_ylabel('Density', fontsize=15)
ax[3].set_ylabel('Density', fontsize=15)
ax[4].set_ylabel('Density', fontsize=15)

ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[3].legend()
ax[4].legend()

ax[0].set_title('Density evolution', fontsize=17)
```
