```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Load data (no gravity)

```python
path = './DataNonGrav/'
data0 = pd.read_csv(path+"nonGrav0.csv")
data1 = pd.read_csv(path+"nonGrav500.csv")
data2 = pd.read_csv(path+"nonGrav1000.csv")
data3 = pd.read_csv(path+"nonGrav1500.csv")
data4 = pd.read_csv(path+"nonGrav2000.csv")
```

### Plotting the jet

```python
import matplotlib as mpl
import matplotlib
fig,ax = plt.subplots(5,1,figsize=(16,12))


cmap = mpl.cm.jet
scat0 = ax[0].scatter(data0['Points:1'], data0['Points:0'], s = 1,
                  cmap = cmap, c = data0['rho'], norm=matplotlib.colors.LogNorm(vmin=7.4*10**(-3), vmax=2.3*10**(2)))
ax[0].tick_params(axis="x",direction="in",pad=-20)
ax[0].set_xticks([5,10,15,20,25,30,35,40,45])
ax[0].tick_params(labelsize=15)

#plt.colorbar(scat0, ax=ax[0])

scat1 = ax[1].scatter(data1['Points:1'], data1['Points:0'], s = 1,
                  cmap = cmap, c = data1['rho'], norm=matplotlib.colors.LogNorm(vmin=7.4*10**(-3), vmax=2.3*10**(2)))
ax[1].tick_params(axis="x",direction="in",pad=-20)
ax[1].set_xticks([5,10,15,20,25,30,35,40,45])
ax[1].tick_params(labelsize=15)
#plt.colorbar(scat1, ax=ax[1])

scat2 = ax[2].scatter(data2['Points:1'], data2['Points:0'], s = 1,
                  cmap = cmap, c = data2['rho'], norm=matplotlib.colors.LogNorm(vmin=7.4*10**(-3), vmax=2.3*10**(2)))
ax[2].tick_params(axis="x",direction="in",pad=-20)
ax[2].set_xticks([5,10,15,20,25,30,35,40,45])
ax[2].tick_params(labelsize=15)
#plt.colorbar(scat2, ax=ax[2])

scat3 = ax[3].scatter(data3['Points:1'], data3['Points:0'], s = 1,
                  cmap = cmap, c = data3['rho'], norm=matplotlib.colors.LogNorm(vmin=7.4*10**(-3), vmax=2.3*10**(2)))
ax[3].tick_params(axis="x",direction="in",pad=-20)
ax[3].set_xticks([5,10,15,20,25,30,35,40,45])
ax[3].tick_params(labelsize=15)
#plt.colorbar(scat1, ax=ax[3])

scat4 = ax[4].scatter(data4['Points:1'], data4['Points:0'], s = 1,
                  cmap = cmap, c = data4['rho'], norm=matplotlib.colors.LogNorm(vmin=7.4*10**(-3), vmax=2.3*10**(2)))
ax[4].set_xticks([5,10,15,20,25,30,35,40,45])
ax[4].tick_params(labelsize=15)
#plt.colorbar(scat2, ax=ax[4])

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.82, 0.15, 0.02, 0.7])
cb = fig.colorbar(scat0, cax=cbar_ax)
cb.set_label(label=r'$\rho [\rho_0]$', size=15)
cb.ax.tick_params(labelsize=14)
    
ax[0].text(42,4.5,'t = 0 years',c='white', fontsize=15)
ax[1].text(42,4.5,'t = 500 years', c='white', fontsize=15)
ax[2].text(42,4.5,'t = 1000 years',c='white', fontsize=15)
ax[3].text(42,4.5,'t = 1500 years', c='white', fontsize=15)
ax[4].text(42,4.5,'t = 2000 years', c='white', fontsize=15)

for i in range(0, 5):
    ax[i].set_ylim(-6.25, 6.25)
    ax[i].set_xlim(0,50)
ax[4].set_xlabel('x $[10^3 AU]$', fontsize=20)
ax[2].set_ylabel('y $[10^3 AU]$', fontsize=20)

plt.subplots_adjust(hspace=0)
plt.show()
```

### Plotting the density at y = 0

```python
_,ax=plt.subplots(5,1,figsize=(8,8))

data0y0 = data0[data0['Points:0'] == 0]
data0y0 = data0y0.sort_values('Points:1')
ax[0].plot(data0y0['Points:1'], data0y0['rho'], c = 'black', label = 't = 0 years')
ax[0].set_xticks([])

data1y0 = data1[data1['Points:0'] == 0]
data1y0 = data1y0.sort_values('Points:1')
ax[1].plot(data1y0['Points:1'], data1y0['rho'], c = 'black', label = 't = 500 years')
ax[1].set_xticks([])

data2y0 = data2[data2['Points:0'] == 0]
data2y0 = data2y0.sort_values('Points:1')
ax[2].plot(data2y0['Points:1'], data2y0['rho'], c = 'black', label = 't = 1000 years')
ax[2].set_xticks([])

data3y0 = data3[data3['Points:0'] == 0]
data3y0 = data3y0.sort_values('Points:1')
ax[3].plot(data3y0['Points:1'], data3y0['rho'], c = 'black', label = 't = 1500 years')
ax[3].set_xticks([])

data4y0 = data4[data4['Points:0'] == 0]
data4y0 = data4y0.sort_values('Points:1')
ax[4].plot(data4y0['Points:1'], data4y0['rho'], c = 'black', label = 't = 2000 years')
ax[4].set_xticks([0,5,10,15,20,25,30,35,40,45,50])


ax[4].set_xlabel(r'$r\ [10^3\ AU]$', fontsize = 15)
ax[2].set_ylabel(r'$\rho\ [\rho_0]$', fontsize = 15)

for i in range(0,5):
    ax[i].legend(fontsize = 15)
    ax[i].tick_params(labelsize = 12)

plt.subplots_adjust(hspace=0)

plt.show()
```
